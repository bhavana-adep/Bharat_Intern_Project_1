#**Model Selection**

```python
df['date']=pd.to_datetime(df['date'])
```

```python
x=df.drop(['price','date'],axis=1)
y=df['price']
```

```python
kfold=KFold(n_splits=5)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=53)
```

#**Model Building**

```python
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
y_pred1=dt.predict(x_test)
score_tree=(100*r2_score(y_test,y_pred1))
score_tree
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred2=lr.predict(x_test)
score_linear=(100*r2_score(y_test,y_pred2))
score_linear
```

```python
lasso=Lasso()
lasso.fit(x_train,y_train)
y_pred3=lasso.predict(x_test)
score_lasso=(100*r2_score(y_test,y_pred3))
```

```python
ridge=Ridge()
ridge.fit(x_train,y_train)
y_pred4=ridge.predict(x_test)
score_ridge=(100*r2_score(y_test,y_pred4))
score_ridge
```

```python
random=RandomForestRegressor()
random.fit(x_train,y_train)
y_pred6=random.predict(x_test)
score_random=(100*r2_score(y_test,y_pred6))
score_random
```

```python
plt.figure(figsize=(9,5))
plt.title("R2_score of Models")
plt.xlabel('Models')
plt.ylabel('r2_score')
plt.bar(['DecisionTree','LinearRegression','Lasso','Ridge','RandomForest'],[score_tree,score_linear,score_lasso,score_ridge,score_random])
plt.show()
```

**Result:** The more accuracy we got from RandomForestRegressor i.e. 88.1%.
